```python
import os
import cv2
import cv2
import numpy as np
import matplotlib as plt
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
from sklearn.linear_model import LinearRegression
from pylab import xticks,yticks,np
from scipy.interpolate import make_interp_spline
```

```python
x_data=pd.read_csv('Hexagon.csv')
x_data=np.array(x_data)
plt.figure(figsize=(25, 4))

plt.grid(True)#生成网格
plt.plot(x_data[:,0],x_data[:,1:])
xticks(np.linspace(0,150,30,endpoint=True))
plt.show()
```

```python
x_data=x_data[:,1:900]
x_data.shape
```

```python
x_data[5]
```

```python
y_data=pd.read_csv('marker_position(2).csv')
```

```python
y_data=y_data[0:900]
y_data=np.array(y_data)
print(y_data.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_data,y_data,test_size=0.4)
```

```python
x_train.shape
```

```python
y_train.shape
```

## 线性回归

```python
clf = LinearRegression()
clf =MultiOutputRegressor(clf)
rf = clf.fit (x_train, y_train)
y_pred = rf.predict(x_test)
print("LinearRegression结果如下：")
print("训练集分数：",rf.score(x_train,y_train))
print("验证集分数：",rf.score(x_test,y_test))
```

## 神经网络

```python
from sklearn.neural_network import MLPRegressor
clf = MLPRegressor()
clf =MultiOutputRegressor(clf)
rf = clf.fit (x_train, y_train)
y_pred = rf.predict(x_test)
print("MLPRegressor结果如下：")
print("训练集分数：",rf.score(x_train,y_train))
print("验证集分数：",rf.score(x_test,y_test))
```

## 梯度提升树

```python
from sklearn.ensemble import GradientBoostingRegressor
clf = GradientBoostingRegressor()
clf =MultiOutputRegressor(clf)
rf = clf.fit (x_train, y_train)
print(x_test.shape)
y_pred = rf.predict(x_test)
print("GradientBoostingRegressor结果如下：")
print("训练集分数：",rf.score(x_train,y_train))
print("验证集分数：",rf.score(x_test,y_test))
```

## bagging

```python
from sklearn.ensemble import BaggingRegressor
clf = BaggingRegressor()
clf =MultiOutputRegressor(clf)
rf = clf.fit (x_train, y_train)
y_pred = rf.predict(x_test)
print("BaggingRegressor结果如下：")
print("训练集分数：",rf.score(x_train,y_train))
print("验证集分数：",rf.score(x_test,y_test))
```
